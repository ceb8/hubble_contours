# Hubble Contours: Basic Usage Example

```python
from hubble_contours import ContourProducer

from IPython.display import Image
```

```python
my_cp = ContourProducer(path=".", RA_DEC_list="SDSS-cross-match-with-mag-and-z")
```

We don't want this example to take too long, so we will cut down the ra/dec list to one.

```python
my_cp.coordlist = my_cp.coordlist[:1]
my_cp.coordlist
```

```python
countour_imgs = my_cp.make_all_contours()
```
